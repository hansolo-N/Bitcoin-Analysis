### Retail Analysis 

```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as datetime
```

```python
df = pd.read_csv('./btc_1d_data_2018_to_2024-09-06.csv',header=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Open time'] = pd.to_datetime(df['Open time'])
df['Close time'] = pd.to_datetime(df['Close time'])
```

```python
df.describe()
```

```python
df['year'] = df['Open time'].dt.year
```

```python
df.head()
```

```python
#
df = df[df['year'] != 2024]
```

```python
print(df['year'].value_counts())
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Open time'], df['Open'], label='Open Price', marker='o')
plt.plot(df['Open time'], df['Close'], label='Close Price', marker='x')
plt.plot(df['Open time'], df['High'], label='High Price', linestyle='--')
plt.plot(df['Open time'], df['Low'], label='Low Price', linestyle='--')
```

```python
plt.title('Bitcoin Price Movement Over Time')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Open time'], df['Volume'], color='purple')
plt.title('Bitcoin Trading Volume Over Time')
plt.xlabel('Time')
plt.ylabel('Volume (BTC)')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df[['Open', 'High', 'Low', 'Close', 'Volume', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume']].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Variables')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Open time'], df['Taker buy base asset volume'], label='Taker Buy Base Asset Volume (BTC)', marker='o')
plt.plot(df['Open time'], df['Taker buy quote asset volume'], label='Taker Buy Quote Asset Volume (USD)', marker='x')

plt.title('Taker Buy Volume (BTC and USD) Over Time')
plt.xlabel('Time')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```
